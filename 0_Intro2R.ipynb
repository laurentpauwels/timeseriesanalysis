### Loops

```r
n=20
for(i in 1:n){ print(i) }
```

#### Using this method, we could create an identity matrix

```r
X = matrix(0, ncol=10,nrow=10)
X # prints out a 10 by 10 matrix of zeros
for(i in 1:10){ X[i,i] = 1 }
X # prints out a 10 by 10 identity matrix

```

In practice, there’s an easier way to create an identity matrix: `diag(1,10)`. We’ll illustrate the central limit using loops. In this experiment, we’re rolling 10 dice (n.dice=10) and recording the results: `round(runif(n.dice,0,5),0)+1`. Each time we roll the 10 dice, we find the average,
`experiment.means[i] = mean(random.data)`.

```r
n.dice = 5 # number of dice in each roll
n = 50 # number of experiments
round(runif(n.dice,0,5),0)+1 ## one roll of 5 dice 
experiment.means = NA # initialise/wipe the object
for(i in 1:n){
  experiment.data = round(runif(n.dice,0,5),0)+1
  experiment.means[i] = mean(experiment.data)
}
hist(experiment.means)
```

### Functions
- We can rewrite the previous experiment in a function, so it’s easy to reuse:

```r
experiment.fn = function(n.dice,n){
  experiment.means = NA # initialise/wipe the object
  for(i in 1:n){
    experiment.data = round(runif(n.dice,0,5),0)+1
    experiment.means[i] = mean(experiment.data)
  }
  hist(experiment.means)
}
# So you can now call the function using:
experiment.fn(n.dice = 5, n = 50) # 5 dice and 50 experiments
experiment.fn(n.dice = 5, n = 100) # 5 dice and 100 experiments
experiment.fn(n.dice = 10, n = 20) # 10 dice and 20 experiments
experiment.fn(n.dice = 20, n = 200) # 10 dice and 20 experiments
# Using functions, it is easy to change the number of trials per experiment and and the number of experiments.
```
